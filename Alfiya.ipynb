```python
def speed(s, isbdy):
    if isbdy:
        s = s - 5
    if   s > 80:
        return 'Big Ticket'
    elif s > 60:
        return 'Small Ticket'
    else:
        return 'No Ticket'
```

```python
speed(81,True)
```

```python
speed(81,False)
```
